```python
split_year=df['released'].astype(str).str.split(',', n= 4, expand=True).rename(columns={0:"Month_date", 1:"year_correct"})
split_year

#df['year_correct']=df['released'].astype(str).str[',':4]
```

```python
df[['Month_date','year_correct']] = df['released'].str.split(',', n=4, expand=True)
df
```

```python
df['year_correct'] = df['year_correct'].str.replace(' ', ' ')
df
```

```python
df['year_correct'] = df.year_correct.str.replace(r"\D+",'')
df
```

```python
df_numerized = df


for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head(10)
```

```python
df_numerized.corr(method='pearson')
```

```python
correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Movies")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
correlation_matrix=df_numerized.corr()
correlation_pairs=correlation_matrix.unstack()
correlation_pairs.head(10)
```

```python
sorted_pairs=correlation_pairs.sort_values()
sorted_pairs.head(10)
```
